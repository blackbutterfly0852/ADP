```python
import pandas as pd 
df_sejong = pd.read_csv("data/6.세종시_오피스텔(매매)_실거래가.csv",encoding='utf8')
```

```python
df_sejong
```

```python
from datetime import datetime
df_sejong.info()
```

```python
### 아래와 같이 계약년월을 datetime으로 변경하세요. 
df_sejong['계약년월'] = df_sejong['계약년월'].astype('str')
df_sejong['계약일'] = df_sejong['계약일'].astype('str').apply(lambda x : x if len(x) == 2 else '0' + x)
df_sejong['계약년월'] = df_sejong['계약년월'] + df_sejong['계약일']
df_sejong['계약년월']
```

```python
#datetime.strptime(df_sejong[['계약년월']], '%Y-%m-%d') # strptime() argument 1 must be str, not DataFrame > 'P'문'날'
```

```python
# 문자열 -> 날짜
df_sejong['계약년월'] = pd.to_datetime(df_sejong['계약년월'], format = '%Y-%m-%d')
df_sejong
```

```python
## 계약년월을 아래와 같이 문자열로 다시 변경하세요. 
```

```python
# 날짜 -> 문자열
df_sejong['계약년월'] = df_sejong['계약년월'].dt.strftime('%Y_%m_%d')
#df_sejong['계약년월'] = df_sejong['계약년월'].str.replace('-','_')
df_sejong
```

```python
## 6월 데이터만 추출해보세요 
df_ymd = df_sejong['계약년월'].str.split('_', expand = True)
df_ymd.columns = ['년','월','일']
df_sejong = pd.concat([df_sejong, df_ymd], axis = 1)
```
