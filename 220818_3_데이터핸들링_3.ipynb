```python
import pandas as pd
import numpy as np
```

```python
ds = pd.read_csv('data/7.세종시_아파트(전월세)_실거래가.csv')
```

```python
ds
```

```python
ds.drop(columns = 'Unnamed: 0', inplace = True)
```

```python
ds.시군구.unique()
```

```python
ds['시군구'].str[0:7]
```

```python
ds['시'] = ds['시군구'].str[0:7]
```

```python
ds
```

```python
# 분할
```

```python
ds['시군구'].str.split(" ", expand = True) # expand = True, 데이터 프레임으로 만든다
```

```python
dss = ds['시군구'].str.split(" ", expand = True)
```

```python
dss
```

```python
dss.columns = ['시', '군', '구']
```

```python
dss
```

```python
ds['군'] = dss['군']
ds['구'] = dss['구']
```

```python
ds
```

```python
# None == None은 False, 비교할 값이 없기 때문
```

```python
ds[ds['구'] == ds['구']] # boolean selection
```

```python
# 가락마을만 조회
ds['단지명'].str.startswith('가락마을') 
# boolean selection? 이용해서 df에서 조회해보자
```

```python
ds[ds['단지명'].str.startswith('가락마을')] # 중간이나 마지막에 '가락마을'이 있다면?
```

```python
ds[ds['단지명'].str.contains('가락마을')]
```

```python
print('startWith : ', len(ds[ds['단지명'].str.startswith('가락마을')]))
print('contains : ', len(ds[ds['단지명'].str.contains('가락마을')]))
```

```python
ds[ds['단지명'].str.contains('가락마을')]['단지명'].unique()
```

```python
# 위의 '이지더원(가락마을10단지)'는 규칙위반으로 보임
# ds.loc[행/boolean selection, 열/:]
ds.loc[ds['단지명']=='이지더원(가락마을10단지)', '단지명'] = '가락마을10단지(이지더원)'
ds.loc[ds['단지명']=='이지더원(가락마을4단지)', '단지명'] = '가락마을4단지(이지더원)'
```
