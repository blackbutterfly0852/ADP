# 과제 3-1 Groupby & crosstab

```python
import pandas as pd
df = pd.read_csv('data/7.세종시_아파트(전월세)_실거래가.csv')
```

## 데이터 : "7.세종시_아파트(전월세)_실거래가.csv"
## 전세의 보증금 평균 값을 구하세요 groupby 사용 


```python
df.columns
```

```python
df['전월세구분'].unique()
df2 = df[df['전월세구분'] == '전세']
df2['보증금(만원)'].mean()

df.groupby('전월세구분').mean()
df.groupby('전월세구분').mean()[['보증금(만원)']].loc[['전세']]
```

## 세종시 월세 평균 가격을 구하세요 


```python
df.head(5)
df.groupby('전월세구분').mean()[['월세(만원)']].loc[['월세']]
```

## 계약년도별, 시군구별 평균 보증금 가격를 구하세요  



```python
df.head(5)
df.groupby(['계약년월','시군구']).mean()[['보증금(만원)']]
```

## 계약년도별 전세 건수를 구하세요 

```python
df.head(10)
```

```python
# 계약년월 별 전세 건수
df2 = df[df['전월세구분']=='전세'].loc[:,['계약년월']]
df2['count'] = 1
df2.groupby('계약년월').count()
```

```python
# 계약년월 별 전세건수_2
pd.crosstab(df['계약년월'], df['전월세구분'], dropna = True )
```

```python
# 계약년도 별 전세건수
df.info() # 계약년월 int 형식.
df['계약년월'] = df['계약년월'].astype('str')
df.info()
```

```python
df['계약년도'] = df['계약년월'].str[0:4]
df['시'] = df['시군구'].str.split(' ', expand=True)[0]
df
```

```python
ct = pd.crosstab(df['계약년도'], df['전월세구분'])
ct
```
