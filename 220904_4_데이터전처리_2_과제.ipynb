```python
# X_train과 X_val의 수치형 변수들만 StandardScaler를 이용하여 표준화 시키세요. -> Z_score 변경
# 정규화 : 데이터셋의 numerical value 범위의 차이를 왜곡하지 않고 공통 척도로 변경하는 것
# 표준화 : 표준정규분포의 속성을 갖도록 피처가 재조정되는 것
# https://deepinsight.tistory.com/165 - fit_transform(train) vs transform(test) 
# fit : train data 학습 후 mean, val 추출
# transform : 위에서 나온 mean, val 값을 바탕으로 test 데이터를 표준화
# fit_transform : train 기준으로, train 데이터의 표준화(test는 사용 X)
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import MinMaxScaler
X_train.head(5)
scaler = StandardScaler()
X_train.info()
```

```python
numerical_columns = ['총구매액', '최대구매액', '환불금액','내점일수', '내점당구매건수', '주말방문비율','구매주기']
x_train_scale = scaler.fit_transform(X_train.loc[:, numerical_columns])
x_val_scale = scaler.transform(X_val.loc[:, numerical_columns])
```

```python
x_train_df = pd.DataFrame(x_train_scale, columns = numerical_columns)
x_val_df = pd.DataFrame(x_val_scale, columns = numerical_columns)
```
