```python
import pandas as pd
```

```python
s = pd.read_csv('data/7.세종시_아파트(전월세)_실거래가.csv')
s
```

```python
# 1) 단지명 중  '해밀마을 2단지'인 row 삭제  
s.drop(s[s['단지명'] == '해밀마을 2단지'].index, inplace = True)
# s[s['단지명']!='해밀마을 2단지'] # 실 데이터는 삭제 X
# s = s.loc[df.단지명!='해밀마을 2단지'] # 실 데이터는 삭제 X
```

```python
s
```

```python
# 2) 'new_시군구' 컬럼을 만들고 '세종특별자치시 고운동' => '세종특별자치시 해밀동' 으로 변경
s['new_시군구'] = s['시군구']
s['new_시군구'].value_counts()
```

```python
s.loc[s['new_시군구'] == '세종특별자치시 고운동', 'new_시군구'] = '세종특별자치시 해밀동'
```

```python
s['new_시군구'].value_counts()
```

```python
# 3) 'Class'컬럼을 만든 뒤 전세면 True 아니면 False로 입력 (loc사용) 
s.columns
```

```python
s['Class'] = False
s.loc[s['전월세구분'] == '전세', 'Class'] = True
# np.where(s['전월세구분'] == '전세', True, False)
```

```python
s.head(5)
```

```python
# 4) 전세이면서 보증금이 1억 이상인 데이터만을 추출 
s.loc[(s['전월세구분'] == '전세') & (s['보증금(만원)'] >= 10000)]
```

```python
# 5) -10~0층 : 지하 , 1~5층 : 저층 6~15층 : 중층, 16층 이상 : 고층으로 할당하는 새로운 컬럼 생성 
import numpy as np
c1 = [(s['층'] <= 0), (s['층']<=5), (s['층']<=15), (s['층']>=16) ]
c2 = ['지하', '저층', '중층', '고층']
s['new_층'] = np.select(c1,c2, default = 'Not Specified')
```

```python
s['new_층'].value_counts()
```

```python
s[s['new_층'] == 'Not Specified'].size
```
