## 결측치 처리
###  결측치 대치법 이해
<b>가.</b> 단순 대치법
* complete analysis : 결측값이 존재하는 행 삭제
* 평균 대치법 : 관측 또는 실험을 통해 얻어진 데이터의 평균으로 결측치 대치
* 단순확률 대치법 : Hot-deck 방법, nearest neighbor 방법 등
    
<b>나.</b> 다중 대치법
* 단순 대치법을 m번 수행해 m개의 가상적 완전 자료를 만듦
* [대치] - [분석] - [결합] 단계

```python
# 이상치 인식 -> 평균/표준편차로 인식, 분산(평균에서 퍼진정도), 어떤 특정 데이터의 퍼진정도가 분산을 넘어선다? 이상치 > boxPlot으로 볼 수 있다.
# Q1-1.5IQR > || Q3-1.5IQ <
```

```python
# IQR을 활용한 이상치 탐색 함수
def detect_outliers(df=None, column=None, weight = 1.5):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    
    IQR = Q3-Q1
    IQR_weight = IQR*weight
    
    outlier_idx = df[(df[column] < Q1 - IQR_weight) | (df[column] > Q3 + IQR_weight)].index
    
    return outlier_idx
```

```python
out_index = detect_outliers(df=air, column='Ozone')
out_index # 61, 116 idx가 이상치로 탐색
```

```python
air.loc[out_index,'outlier'] = True # outlier 컬럼 추가 후 True값
```

```python
air[air['outlier'] == True]
```

```python
# Outlier 대체 (중위 값 혹은 평균값으로) : 보통의 경우 관측이 잘못되었을 때
# Outlier 자체가 의미있을 수 있다(불량탐지)
air.loc[out_index,'Ozone'] = air['Ozone'].median()
air.loc[out_index,'Ozone']
```

```python
sns.boxplot(x='Ozone', data=air) # 이상치를 대치하고 boxplot을 그리면 재계산되서 이상치가 나올 수 있다.
```
