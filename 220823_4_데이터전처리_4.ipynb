```python
# 범주형 데이터 전처리하기
# 연속형 변수 : 수치형변수
# 범주형 변수 : 이산형 혹은 category 등 -> 수치형이지만 내부 데이터를 살펴보고 범주형인지 파악 필요
# 범주형 변수를 표현하는데 가장 널리 쓰이는 방법 -
# > 원-핫-인코딩
# 범주형 변수에 있는 원소를 columns로 변경 -> 해당 범주에 속하면 1, 아니면 0으로 채우는 기법
# pd.get_dummies('범주형변수')
# pd.get_dummies(data, columns=['범주형1','범주형2'])
```

```python
df = pd.read_csv('data\빅분기_X_train.csv', encoding = 'cp949')
```

```python
df
```

```python
# 원-핫-인코딩 : 머신러닝에 명목형 변수를 대입시 오류 발생 때문
df['주구매상품'].value_counts()
df['주구매지점'].value_counts()
```

```python
print(len(df['주구매상품'].unique()))
print(len(pd.get_dummies(df['주구매상품']).columns))
```

```python
# 원본 데이터에서 범주형 변수를 대체하기
df_num = pd.get_dummies(df, columns=['주구매상품','주구매지점'])
```

```python
df_num
```

```python
# 위 방법을 사용하지 않는다면 코드도 복잡하며, 추가된 column도 직관적으로 알기 어려움.
pd.concat([df, pd.get_dummies(df['주구매상품']), pd.get_dummies(df['주구매상품'])], axis=1).drop(columns=['주구매지점','주구매상품'])
```
