```python
# fifa 데이터에서 passing과 overall 사이의 선형회귀모형 생성 뒤, 직선 그래프 표현
# 선형 회귀식에 대한 그래프는 passing과 overall에 대한 산점도 위에 나타내며
# 색상은 빨강으로 지정
# overall의 중위수는 수평선, passing의 중위수는 수직선
# 각각의 선은 점 스타일의 회색 선으로 지정
```

```python
fifa = pd.read_csv('data/players_20.csv')
```

```python
fifa
```

```python
feature = ['sofifa_id', 'short_name', 'passing', 'overall']
fifa = fifa[feature].head(100)
fifa.dropna(inplace = True)
fifa.head()
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression() # 회귀모델 선언
x = fifa[['passing']]
lr.fit(x, fifa['overall']) # 회귀모델 생성 (passing을 기준으로 overall 예측)

plt.scatter(fifa['passing'], fifa['overall']) # 원 데이터 산포도
plt.plot(fifa['passing'], lr.predict(x), c = 'r', linestyle='-') # 회귀직선
# 수평선 : overall의 중위값, 범위는 passing의 mix,max
plt.hlines(np.median(fifa['overall']), np.min(fifa['passing']), np.max(fifa['passing']), 'lightgrey' , linestyles = ':')
# 수직선 : passing의 중위값, 범위는 overall의 mix,max
plt.vlines(np.median(fifa['passing']), np.min(fifa['overall']), np.max(fifa['overall']), 'lightgrey' , linestyles = ':')


plt.xlabel('passing', fontsize=12)
plt.ylabel('overall', fontsize=12)
plt.title('fifa data')
plt.show()
```

```python
# 곡선 그래프
# 평균이 0이고 표준편차가 1인 확률밀도함수를 생성하고 그래프로 그리기
# x축의 범위는 -3 + 3으로 한다
```

```python
import matplotlib.pyplot as plt
import numpy as np
import scipy.stats as stats
```

```python
# np.linespace(min,max,cnt) : 지정된 가격 동안 균등 한 간격의 숫자를 반환
# -3에서 3까지 100등분하여 값을 지정함
x = np.linspace(-3,3,100)
# stats.norm(mean,sd) : 확률밀도를 계산해주는 함수 (평균 0, 표준편차 1)
y = stats.norm(0,1).pdf(x)

plt.plot(x,y)
plt.xlabel('x') # x축 레이블 지정
plt.ylabel('y') # y축 레이블 지정
plt.grid() # 플롯에 격자 보이기
plt.title("Nomal Distribution withdout scipy")
plt.show()
```

```python
# np.linespace(min,max,cnt) : 지정된 가격 동안 균등 한 간격의 숫자를 반환
# -3에서 3까지 100등분하여 값을 지정함
x = np.linspace(-3,3,100)
# stats.norm(mean,sd) : 확률밀도를 계산해주는 함수 (평균 0, 표준편차 1)
y = stats.norm(0,1).pdf(x)

plt.scatter(x,y) # 산점도로 그리기
plt.xlabel('x') # x축 레이블 지정
plt.ylabel('y') # y축 레이블 지정
plt.grid() # 플롯에 격자 보이기
plt.title("Nomal Distribution withdout scipy")
plt.show()
```

```python
cars = pd.read_csv('data/Cars93.csv')
cars.head()
```

```python
# Cars93에서 Wheelbase 따른 Type 별 Price 관측값을 살펴보자
cars.Wheelbase.value_counts()
```

```python
# 미 Sort
car_type = cars.Type.unique()
print('car_type : ' , car_type)

plt.plot('Wheelbase', 'Price', data=cars.loc[cars['Type']=='Small'])
plt.plot('Wheelbase', 'Price', data=cars.loc[cars['Type']=='Midsize'])

plt.legend(car_type)
plt.show()
```

```python
# Sort
car_type = cars.Type.unique()
print('car_type : ' , car_type)
car_sorted = cars.sort_values(by=['Wheelbase'])
plt.plot('Wheelbase', 'Price', data=car_sorted.loc[car_sorted['Type']=='Small'])
plt.plot('Wheelbase', 'Price', data=car_sorted.loc[car_sorted['Type']=='Midsize'])

plt.legend(car_type)
plt.show()
```
