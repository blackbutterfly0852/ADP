```python
# 조건이 여러개인 경우
# &가 |보다 우선순위를 갖는다. 조건연산자의 우선순위를 모르는 경우 괄호로 묵는다.
# 아래에서 (df_t['fare'] > 30) & (df_t['class'] == "First") 먼저 실행
```

```python
df_t[df_t['survived']==1 | (df_t['fare'] > 30) & (df_t['class'] == "First")]
```

```python
# 조건에 따른 column 추가, 변경
```

```python
df_t.loc[df_t['survived'] == 1, 'temp'] = 99999
df_t
```

```python
df_t['temp2'] = False
df_t.loc[df_t.age > 30 , 'temp2']
```

```python
df_t.loc[df_t.age > 30, 'temp2'] = '30대 이상입니다.'
```

```python
# loc를 사용하지 않을 경우 오류
# a copy of a slice from a DataFrame : 메모리에만 올라가 있는 복사본이기 때문이다.
df_t[df_t['age'] > 30]['temp2'] = '30대 이상입니다.'
```

```python
df_t.drop(columns = ['temp', 'temp2'], inplace = True)
```

```python
df_t
```

```python
# np.where()
# 데이터 프레임 or 리스트내 필드에 접근하고 수정하는 방법(loc, iloc를 두번 사용하는 것과 동일)
# np.where(condition, x,y)는 조건이 충족되면 x를 그렇지 않으면 y를 반환
```

```python
df_t['alive'] = np.where(df_t['survived']==1, 'Yes', 'No')
```

```python
df_t
```

```python
# np.select()
# 범주형 변수를 만들때 사용
# 데이터 프레임 or 리스트 내 필드에 접근하고 수정하는 방법.
# 조건 목록과 선택목록을 입력으로 받아서, 조건에 따라 선택 목록 요소로 구성된 배열을 반환합니다.
# 이 방법을 사용하면 두 개 이상의 조건이 있었을 때 Pandas의 주어진 조건을 기반으로 DataFrame 열을 만들 수 있다.
```

```python
cl = [
    (df_t['age'] >= 19),
    (df_t['age'] >= 8 & (df_t['age'] < 19)),
     df_t['age'] < 8
]
chl = [
    '성인',
    '초중고등학생',
    '미취학아동'
]

df_t['status'] = np.select(cl,chl, default ='Not_specified')
```
