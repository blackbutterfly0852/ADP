```python
# 데이터 요약(중요)
# 특정 column을 기준으로 데이터를 그룹지어 집계함수 적용
# df.groupby("그룹화할 기준열").function()
# df.groupby(["컬럼명1", "컬럼명2"]).function() : 2개 이상의 그룹화
```

```python
import pandas as pd

di = pd.read_csv("data/iris.csv")
```

```python
di
```

```python
di.groupby(["target"]) # 객체 생성, 그룹화만 진행
```

```python
di.groupby("target").mean() #groupby 내 명목형, 범주형, 카테고리
```

```python
di.groupby(["sepal length"]).mean() # 집계가 될 수 없는 명목형 변수는 사라짐
```

```python
di.groupby(["target"]).mean()
```

```python
 di.groupby(["target"]).mean().reset_index()
```

```python
di.groupby(["target"]).mean()[["sepal width"]]
```

```python
di.groupby(["target"]).mean()[['sepal width', 'petal width']]
```

```python
# Titanic에서 좌석등급과 생존여부의 관계를 살펴보기 위해 Class 변수에 따른 Survived 변수의 도수를 표 형태로
pd.DataFrame(dt[['class','survived']].value_counts())
```

```python
pd.crosstab(dt['class'], dt['survived'], dropna = True) # EDA : 데이터를 보고 인사이트
```

```python
# 범주형 변수에 대한 상대도수 비율
# pd.crosstab(index, columns, dropna = , normalize = )
# normalize = True 이면 모든 값에 대해 정규화
# normalize = 'index'를 전달하면 행에 대해 정규화
# normalize = 'columns'를 전달하면 열별로 정규화
```

```python
# 성별 변수에 따른 생존여부의 관계를 전체에 대한 비율, 행별비율, 열별 비율로 살피기
```

```python
pd.crosstab(dt['sex'],dt['survived'])
```

```python
pd.crosstab(dt['sex'],dt['survived'], normalize = True)
```

```python
pd.crosstab(dt['sex'],dt['survived'], normalize = 0) # 행 별로 정규화
pd.crosstab(dt['sex'],dt['survived'], normalize = 'index') # 행 별로 정규화
```

```python
pd.crosstab(dt['sex'],dt['survived'], normalize = 1) # 열 별로 정규화
pd.crosstab(dt['sex'],dt['survived'], normalize = 'columns') # 열 별로 정규화
```
