##  players_19.csv와 players_20.csv 데이터를 행결합해서 master data를 만드세요 

```python
import pandas as pd 
```

```python
f19 = pd.read_csv("data/players_19.csv")
f20 = pd.read_csv("data/players_20.csv")
```

```python
f19 = f19.loc[:, ['sofifa_id','short_name','overall']]
f20 = f20.loc[:, ['sofifa_id','overall']]
```

```python
## fifa_df19 = players_19.csv에서 sofifa_id','short_name','overall' 컬럼만 추출
## fifa_df20 = players_20.csv 에서 sofifa_id와 overall만을 추출 
## sofifa_id를 key값으로  위의 두 테이블을 innerjoin 하세요 
```

```python
f19.head(10)
```

```python
f20.head(10)
```

```python
inner = pd.merge(f19, f20, how = 'inner', on ='sofifa_id')
inner
```

## sofifa_id를 key값으로  위의 두 테이블을 outerjoin 하세요 

```python
outer = pd.merge(f19, f20, how = 'outer', on = 'sofifa_id')
outer
```

## outerjoin을 한 테이블 중 inner join을 한 테이블과 중복되는 행을 삭제하세요 

```python
# 시리즈끼리 비교할 때에는 series1.isin(series2)를 사용하시면 됩니다.
# series1의 값이 series2에 있다면 True로, 없으면 False로 return 됩니다.
# f22 : outer, f21 : inner
isIn = outer['sofifa_id'].isin(inner['sofifa_id'])
outer['compare'] = isIn
outer
```

```python
# outer에서 삭제방법 1 (outer에서 실 데이터 삭제)
outer = outer.drop(outer[outer['compare'] == True].index)
outer[outer['compare'] == True]
```

```python
# outer에서 삭제방법 1 (outer에서 실 데이터 삭제 X)
#outer[~outer['compare'].isin(inner['compare'])]
outer[~outer['compare'] == True]
```
