```python
import pandas as pd
df = pd.read_csv('data/iris.csv')
df
```

```python
# 산점도
import matplotlib.pyplot as plt
df[['sepal length', 'petal length']] # sepal 꽃받침, petal 꽃잎
print(df['sepal length'].min())
print(df['sepal length'].max())
print(df['petal length'].min())
print(df['petal length'].max())
```

```python
plt.scatter(x=df['sepal length'], y=df[['petal length']], c = 'r', s = 50, alpha = 1)
plt.xlabel('sepal length')
plt.ylabel('petal length')
plt.title('petal length related with petal length')
plt.show()
# 꽃받침의 길이가 길수록, 대부분 꽃잎의 길이가 길다.
```

```python
# seaborn
import seaborn as sns
sns.scatterplot(x = 'sepal length',
                y = 'petal length',
                hue = 'target',
                style = 'target',
                s=50,
                data=df)
```

# Data :  Cars93 데이터 
## X축 : RPM, Y축 Price 산점도 그리기 
## X축 Passengers, Y축 Price, 꺾은선 그래프 그리기 
## Price, RPM 변수 Boxplot 그리기 

```python
df2 = pd.read_csv('data/Cars93.csv')
df2[['RPM', 'Price']]

fig, axs = plt.subplots(2,2,figsize = (15,10))
axs[0][0].scatter(df2['RPM'], df2['Price'])
axs[0][0].set_title('RPM with Price')

df2.sort_values(by = ['Passengers', 'Price'], inplace = True)
axs[0][1].plot(df2['Passengers'], df2['Price'])
axs[0][1].set_title('Passengers with Price')

axs[1][0].boxplot(df2['Price'])
axs[1][0].set_title('Price in Boxplot')

axs[1][1].boxplot(df2['RPM'])
axs[1][1].set_title('RPM in Boxplot')
```
