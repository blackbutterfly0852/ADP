```python
import pandas as pd
```

```python
# 1) 세종시 데이터에서 컬럼 확인 후, unnamed 컬럼 제거
sejong.columns
sejong.drop(columns = ['Unnamed: 0'], inplace = True)
# sejong.drop('Unnamed: 0', axis = 1, inplace = True)
```

```python
# 2) 전월세구분 컬럼에서 unique값 확인
sejong['전월세구분'].unique()
```

```python
# 3) index= 100~200, column = 전월세구분, 전용면적(㎡),보증금(만원)으로 이루어진 데이터프레임 추출 
sejong_2 = sejong.loc[100:200, ['전월세구분', '전용면적(㎡)','보증금(만원)']]
sejong_2
```
